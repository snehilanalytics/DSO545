```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv('mpg.csv')
```

```python
data.head()
```

```python
#Q1 density plot
```

```python
plt.figure(figsize=(12,10)) #Kernel Density Plot

sns.kdeplot(data.loc[data.cyl==4,'hwy'],
            shade=True,
            color="green",
            label="Cyl=4"
           )
sns.kdeplot(data.loc[data.cyl==5,'hwy'],
            shade=True,
            color="deeppink",
            label="Cyl=5",
            alpha=0.4
           )
sns.kdeplot(data.loc[data.cyl==6,'hwy'],
            shade=True,
            color="dodgerblue",
            label="Cyl=6"
           )
sns.kdeplot(data.loc[data.cyl==8,'hwy'],
            shade=True,
            color="orange",
            label="Cyl=8"
           )


plt.show()
```

```python
#Create a 2X3 grid of plots
 #Height space between top and bottom
plt.figure() #2X3 subplot
plt.subplots_adjust(hspace=0.4,wspace=1.6)
for i in range(1,7):
    plt.subplot(2,3,i) #Last argument is an index to refer to a specific plot
    plt.text(0.5,0.5,str((2,3,i)),ha="center") #0.5 for x axis, 0.5 for y and a tuple is passed for string
# ha is horizontal alignment & va for Vertical
plt.show()
```

```python
#plot sine and cosine
```

```python
plt.figure()
fig,ax=plt.subplots(2,3,sharex='col',sharey="row") #Sharex sharing the same x axis
plt.subplots_adjust(hspace=0.4,wspace=0.6)

x=np.linspace(0,50)
ax[0,0].plot(np.sin(x))
ax[0,2].plot(np.cos(x),color='red')



plt.show()
```

```python
#5 Marginal Boxplots
```

```python
fig=plt.figure(figsize=(10,12))
grid=plt.GridSpec(4,4,wspace=0.4,hspace=0.5)
ax_main=fig.add_subplot(grid[0:3,0:3]) #Can use grid[0:-1] #-1 would select one before the end
ax_right=fig.add_subplot(grid[0:3,3])
ax_bottom=fig.add_subplot(grid[3,0:3])
#Plot the main graph 
ax_main.scatter(data.displ,data.hwy)
sns.boxplot(data.hwy,ax=ax_right,orient='v') #Orient gives a vertical orientation to boxplot
ax_right.set(xticks=[],yticks=[],ylabel="") # Remove ticks and labels
sns.boxplot(data.displ,ax=ax_bottom)
ax_bottom.set(xticks=[],yticks=[],xlabel="")
ax_main.set(title='Scatterplot & Boxplots \n Displacement v/s Highway Mileage',xlabel='displ',ylabel='Hwy')
plt.show()
```

```python
#Last Question Secondary Y axis
data2=pd.read_csv('economics.csv')
```

```python
data2.head()
```

```python
fig=plt.figure(figsize=(12,12))
plt.plot(data2.date,data2.psavert,c='tab:red')
#Set the x axis #To show for every 5 years
ax=plt.gca()
#Set the main y axis (left)
ax.tick_params(axis='y',labelcolor='tab:red')
ax.set_ylabel("Personal Savings Rate",c='tab:red',fontsize=14)
ax.set_xticks(np.arange(0,len(data2.date),60))

#Tell matplot to share the y axis, set secondary Y axis
ax_right=ax.twinx() # Create a twin y axis that share the same X axis
ax_right.plot(data2.date,data2.unemploy,c='tab:blue')
ax_right.tick_params(axis='y',labelcolor='tab:blue')
ax_right.set_ylabel("Unemployment Rate",c='tab:blue',fontsize=14)
ax_right.set_xticks(np.arange(0,len(data2.date),60))
plt.title("PSR Vs Unemployment",fontsize=18)

plt.show()
```
